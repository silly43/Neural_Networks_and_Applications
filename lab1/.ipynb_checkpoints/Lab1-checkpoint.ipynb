```python
import pandas as pd

data = pd.read_csv("Concrete_Data_Yeh.csv")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#相关性矩阵
corr = data.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Matrix")
plt.show()
```

- 通过相关性分析可以发现，cement、superplasticizer和age与混凝土抗压强度具有较强的正相关性，而water与强度呈较强的负相关关系，因此这些特征对模型预测具有重要影响。

```python
X = data.iloc[:,0:8].values
y = data.iloc[:,8].values

train_size = int(0.8 * len(X))

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]
```

- 将数据集的前80%作为训练集，后20%作为测试集

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #创建一个标准化器

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

- 计算训练集每个特征的均值和标准差，并对训练集和测试集数据进行标准化

```python
import torch
import torch.nn as nn
import torch.optim as optim

# 转换为tensor
X_train = torch.FloatTensor(X_train)
y_train = torch.FloatTensor(y_train).view(-1,1)

X_test = torch.FloatTensor(X_test)
y_test = torch.FloatTensor(y_test).view(-1,1)

# 神经网络
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.net = nn.Sequential(
            nn.Linear(8,16),
            nn.ReLU(),
            nn.Linear(16,8),
            nn.ReLU(),
            nn.Linear(8,1)
        )

    def forward(self,x):
        return self.net(x)

model = Net()
```

```python
if hasattr(model, 'loss_curve_'):
    losses = model.loss_curve_
    iterations = range(1, len(losses) + 1)
    
    # 绘制损失曲线
    plt.figure(figsize=(12, 5))
    
    # 子图1：损失曲线
    plt.subplot(1, 2, 1)
    plt.plot(iterations, losses, 'b-', linewidth=2)
    plt.xlabel('迭代次数', fontsize=12)
    plt.ylabel('损失值', fontsize=12)
    plt.title('训练损失曲线', fontsize=14)
    plt.grid(True, alpha=0.3)
    
    # 添加最终损失值标注
    final_loss = losses[-1]
    plt.text(iterations[-1], final_loss, f'最终损失: {final_loss:.6f}', 
             fontsize=10, verticalalignment='bottom')
    
    # 子图2：对数坐标下的损失曲线（更好地观察收敛过程）
    plt.subplot(1, 2, 2)
    plt.semilogy(iterations, losses, 'r-', linewidth=2)
    plt.xlabel('迭代次数', fontsize=12)
    plt.ylabel('损失值 (对数坐标)', fontsize=12)
    plt.title('训练损失曲线 (对数坐标)', fontsize=14)
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # 打印训练统计信息
    print(f"\n训练完成！")
    print(f"总迭代次数: {len(losses)}")
    print(f"初始损失: {losses[0]:.6f}")
    print(f"最终损失: {losses[-1]:.6f}")
    print(f"损失降低: {((losses[0] - losses[-1]) / losses[0] * 100):.2f}%")
    
else:
    print("注意：此版本的scikit-learn不提供loss_curve_属性")
    print("训练过程已通过verbose=True在控制台输出")
    
    # 模拟实时可视化（如果需要）
    # 可以创建一个交互式绘图来实时显示损失
    plt.ion()  # 开启交互模式
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # 模拟训练过程（实际应用中需要从训练回调获取数据）
    print("\n提示：对于实时可视化，可以考虑使用自定义训练循环")
    
    plt.ioff()
    plt.show()

# 模型评估
print("\n模型信息：")
print(f"模型架构: {model.hidden_layer_sizes}")
print(f"激活函数: {model.activation}")
print(f"学习率: {model.learning_rate_init}")
print(f"最大迭代次数: {model.max_iter}")
print(f"实际迭代次数: {model.n_iter_}")
print(f"最终损失值: {model.loss_}")
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Test MSE:", mse)
```

```python
plt.figure()
plt.scatter(y_test, y_pred)

plt.xlabel("True Strength")
plt.ylabel("Predicted Strength")

plt.title("Prediction vs True Value")

plt.show()
```
